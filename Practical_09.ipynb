```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure and subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Plot data on each subplot with different colors
axes[0].plot(x, y1, color='blue', label='Sine')
axes[1].plot(x, y2, color='red', label='Cosine')
axes[2].plot(x, y3, color='green', label='Tangent')

# Add titles and labels to each subplot
axes[0].set_title('Sine Function')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].legend()

axes[1].set_title('Cosine Function')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Y-axis')
axes[1].legend()

axes[2].set_title('Tangent Function')
axes[2].set_xlabel('X-axis')
axes[2].set_ylabel('Y-axis')
axes[2].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
