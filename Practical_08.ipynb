```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data for histograms
data = np.random.randn(100)

# Creating a histogram
plt.figure(figsize=(9,6))
plt.hist(data, bins=20,  color='salmon', edgecolor='darkred')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Generating data for sine and cosine functions
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Creating plots for sine and cosine functions
plt.figure(figsize=(9, 6))
plt.plot(x, y_sin, label='Sine', color='darkgreen')
plt.plot(x, y_cos, label='Cosine', color='red')
plt.title('Sine and Cosine Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
